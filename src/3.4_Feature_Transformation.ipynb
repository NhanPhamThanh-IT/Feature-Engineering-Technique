### __Importing the dependencies__

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split

from utils import transformation
```

### __Loading Dataset__

```python
use_cols = [
    'Pclass', 'Sex', 'Age', 'Fare', 'SibSp',
    'Survived'
]

data = pd.read_csv('../datasets/titanic.csv', usecols=use_cols)
```

```python
data.head(3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(data, data.Survived, test_size=0.3,
                                                    random_state=0)
X_train.shape, X_test.shape
```

### __Logarithmic transformation__

```python
X_train_copy = X_train.copy(deep=True)
X_train_copy = transformation.log_transform(data=X_train,cols=['Fare'])
```

```python
print(X_train_copy.head(6))
```

### __Reciprocal transformation__

```python
X_train_copy = X_train.copy(deep=True)
X_train_copy = X_train_copy[X_train_copy.Fare!=0]
X_train_copy = transformation.reciprocal_transform(data=X_train_copy,cols=['Fare'])
```
